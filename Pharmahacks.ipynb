```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# extract position data and timestamps
virmen_positions = nwb.acquisition['Virmen_positions']['Virmen_forward_lateral_positions'].data[:]
virmen_timestamps = nwb.acquisition['Virmen_positions']['Virmen_forward_lateral_positions'].timestamps[:]

print(virmen_positions.shape, virmen_timestamps.shape)
assert virmen_positions.shape[0] == virmen_timestamps.shape[0]

# plot a portion of position data
t_start = 5000
t_end = 10000
fig, ax = plt.subplots(2,1,figsize=(6,4), constrained_layout=True)
ax[0].plot(virmen_timestamps[t_start:t_end], virmen_positions[t_start:t_end,0],'k')
ax[0].set(ylabel='position (m)', title='Forward position')
ax[1].plot(virmen_timestamps[t_start:t_end], virmen_positions[t_start:t_end,1],'k')
ax[1].set(xlabel='Time (sec)', ylabel='position (m)', title='Lateral position');
```

```python
# extract frame-aligned behavioral variables
imaging_frame_timestamps = nwb.processing['behavior']['frame_aligned_position']['frame_aligned_forward_and_lateral_position'].timestamps[:]
frame_aligned_position = nwb.processing['behavior']['frame_aligned_position']['frame_aligned_forward_and_lateral_position'].data[:]
frame_aligned_velocity = nwb.processing['behavior']['frame_aligned_velocity']['frame_aligned_pitch_roll_yaw_velocity'].data[:]
frame_aligned_trial_number = nwb.processing['behavior']['frame_aligned_trial_number'].data[:]
frame_aligned_time_from_choice_point = nwb.processing['behavior']['frame_aligned_time_from_choice_point'].data[:]
plane_idx_for_imaging_frames = nwb.processing['behavior']['plane_idx_for_imaging_frames'].data[:]

# get forward and lateral positions
posF = frame_aligned_position[:,0]
posL = frame_aligned_position[:,1]

# get pitch, roll and yaw velocities
pitch = frame_aligned_velocity[:,0]
roll = frame_aligned_velocity[:,1]
yaw = frame_aligned_velocity[:,2]
```

```python
data = pd.concat([pd.Series(imaging_frame_timestamps), pd.Series(posF), pd.Series(posL)], axis=1)
```

```python
data.columns=["time", "pos_x", "pos_y"]     #pos_x and pos_y as target variables

#replace NAs with 0
data['pos_x'].fillna(0.0, inplace=True)
data['pos_y'].fillna(0.0, inplace=True)

#get data_x and data_y
data_x= data.iloc[:, 0]
data_y = data.iloc[:, -2:]
```

```python
#imports
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

Tests with neuron data

```python
imaging_timestamps_0 = nwb.processing['ophys']['deconvolved_activity_plane_0'].timestamps
deconv_0 = nwb.processing['ophys']['deconvolved_activity_plane_0'].data[:]
df_0 = nwb.processing['ophys']['df_over_f_plane_0']['dF_over_F_plane_0'].data[:]
```

```python
arr1 = np.array(imaging_timestamps_0).flatten()
arr2= np.array(imaging_frame_timestamps).flatten()

#merging the timestamps from neurons data and position data so they are consistent

indices = []
cur_index = 0
for time1 in arr1:
  for idx_time in range(cur_index, len(arr2)):
    if abs(time1-arr2[idx_time]) < 1:
      indices.append(idx_time)
      cur_index = idx_time+1
      break


```

```python
data_new = data.iloc[indices]
```

```python
data_new = pd.concat([pd.DataFrame(deconv_0), data_new]).fillna(0)  #add neuron features
```

```python
data_new.columns = [str(col) for col in data_new.columns]
data_new_x = data_new.iloc[:, 0:-2]
data_new_y = data_new.iloc[:, -2:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_new_x, data_new_y, test_size=0.2, random_state=42)
```

```python
# Cross validation to select best k
from sklearn.model_selection import cross_val_score

knn_regressor = KNeighborsRegressor(n_neighbors=3)

k_values = list(range(1, 7))  # Try k from 1 to 10
cv_scores = []
for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=3, scoring='neg_mean_squared_error')
    cv_scores.append(-np.mean(scores))  # Convert negative MSE to positive

#Choose k
best_k = k_values[np.argmin(cv_scores)]
print("Best k:", best_k)
```

```python
best_k=2
```

```python
#remove unnecessary features
from sklearn.feature_selection import SelectKBest, f_regression

num_features=180

feature_selector = SelectKBest(f_regression, k=num_features)

feature_selector.fit(X_train, np.array(y_train)[:, 0])
scores_1 = feature_selector.scores_

feature_selector.fit(X_train, np.array(y_train)[:, 1])
scores_2 = feature_selector.scores_

scores_combined = scores_1 + scores_2
```

```python
remove_indices = np.argsort(scores_combined)[:num_features]

data_dropped_x = data_new_x.drop(data_new_x.columns[remove_indices], axis=1)
```

```python
data_dropped_x
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_dropped_x, data_new_y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

knn_regressor = KNeighborsRegressor(n_neighbors=best_k)

knn_regressor.fit(X_train, y_train)

y_pred_knn = knn_regressor.predict(X_test)

mse = mean_squared_error(y_test, y_pred_knn)
mae = mean_absolute_error(y_test, y_pred_knn)
r2 = r2_score(y_test, y_pred_knn)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R2 Score:",r2)
```
